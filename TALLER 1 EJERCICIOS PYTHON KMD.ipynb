```python
# EJERCICIO 2
pala = input("Porfavor ingrese la contraseña: ")
a = []  
b = [] 
c = []
caracter_especial = [ "*", "_" , "!", "$", "%" ]
for i in range(len(pala)):
    if pala[i] in caracter_especial:
        c.append(True)
    else:
        c.append(False) #Caracter especial
    a.append((pala[i].isnumeric())) #Caracter numerico
    b.append((pala[i].isupper())) #Mayuscula
if sum(a) == 0 or sum(b) == 0 or sum(c) == 0:
    print("La contraseña es incorrecta. Falta al menos un carácter numérico o una mayúscula o uno de los siguientes carácter especiales : *, _ , !, $, % ")
else:
    print("La contraseña es correcta.")
```

```python
# EJERCICIO 3
lista_nueva = ["Libro_1 = 50000", "Libro_2 = 30000", "Libro_3 = 80000", "Libro_4 = 20000", "Libro_5 = 100000", "Libro_6 = 40000"]
from itertools import combinations
f = []
temp = combinations(lista_nueva, 3)
for i in list(temp):
	f.append(i)
```

```python
f
```

```python
s = []
for i in range(len(f)):
    s.append(int(f[i][len(f[i])-3].split()[len(f[i])-1]) + int(f[i][len(f[i])-2].split()[len(f[i])-1]) +
             int(f[i][len(f[i])-1].split()[len(f[i])-1]))  
s       
```

```python
h = []
for i in range(len(f)):
    if int(f[i][len(f[i])-3].split()[len(f[i])-1]) > int(f[i][len(f[i])-2].split()[len(f[i])-1]):
        if int(f[i][len(f[i])-2].split()[len(f[i])-1]) > int(f[i][len(f[i])-1].split()[len(f[i])-1]):
            h.append(int(f[i][len(f[i])-1].split()[len(f[i])-1]))
        else:
            h.append(int(f[i][len(f[i])-2].split()[len(f[i])-1]))
    else:
        if int(f[i][len(f[i])-3].split()[len(f[i])-1]) > int(f[i][len(f[i])-1].split()[len(f[i])-1]):
            h.append(int(f[i][len(f[i])-1].split()[len(f[i])-1]))
        else:
            h.append(int(f[i][len(f[i])-3].split()[len(f[i])-1]))
h            
```

```python
Total_Pagado = [s-h for s,h in zip(s,h)]
print(Total_Pagado)
```

```python
lista = ["Libro_1", "Libro_2","Libro_3","Libro_4","Libro_5", "Libro_6"]
d = []
temp = combinations(lista, 3)
for i in list(temp):
	d.append(i)
```

```python
user_list = list(zip(d, s, h, Total_Pagado))
import pandas as pd
df = pd.DataFrame(user_list,
                  columns=['Combinacion','Total de libros', 'Descuento de Promoción', 'Total Pagado'])
df
```
